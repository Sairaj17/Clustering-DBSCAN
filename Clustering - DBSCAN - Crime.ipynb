# Perform Clustering(Hierarchical, Kmeans & DBSCAN) for the crime data and identify the number of clusters formed and draw inferences.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
```

```python
crime = pd.read_csv('C:/Users/ssair/datasets/crime_data.csv')
crime
```

```python
crime2=crime.drop(['Unnamed: 0'],axis=1)
crime2
```

# EDA

```python
crime2.info()
```

```python
# Normalize heterogenous numerical data by using Standard Scaler
crime3=StandardScaler().fit_transform(crime2)
```

```python
# DBSCAN Clustering
dbscan=DBSCAN(eps=1,min_samples = 3)
dbscan.fit(crime3)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
# Adding clusters to dataset
crime2['clusters']=dbscan.labels_
crime2
```

```python
crime2.groupby('clusters').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(crime2['clusters'],crime2['Murder'], c=dbscan.labels_)
```
